```python
from google import genai

client  = genai.Client()
```

```python
chat_bot = client.chats.create(model = "gemini-2.5-flash")
```

```python
resposta = chat_bot.send_message("O que é a Inteligência Artificial")
```

```python
print(resposta.text)
```

```python
resposta = chat_bot.send_message("Quando foi criado o Gemini")
print(resposta.text)
```

```python
chat_bot.get_history()
```

```python
prompt = input("digite sua pergunta: ")

while prompt != "fim":
  resposta = chat_bot.send_message(prompt)
  print(resposta.text)
  print("\n")
  prompt = input("digite sua pergunta, ou diga 'fim' para finalizar: ")
```

```python
lista_de_perguntas = [
    "Como funciona um motor V16?",
    "Como a gasolina vai ao motor?",
    "Como funciona a direção hidráulica?",
    "Quantos cavalos tem um Mclaren Senna?",
    "Quanto custa uma Ferrari Monza",
    "Qual o valor do IPVA no Brasil?",
    "Como calcular o gasto do motor?",
    "O que é a Inteligência Artificial",
    "Quando foi criado o Gemini",
    "Quantos anos tem Lebron James?"
]

with open("perguntas.txt", "w", encoding="utf-8") as arquivo:
  for pergunta in lista_de_perguntas:
    arquivo.write(pergunta + "\n")


```

```python
lista_desafio = []

with open ("perguntas.txt", "r", encoding="utf-8") as arquivo:
  for linha in arquivo:
    lista_desafio.append(linha.strip())

print(lista_desafio)
```
